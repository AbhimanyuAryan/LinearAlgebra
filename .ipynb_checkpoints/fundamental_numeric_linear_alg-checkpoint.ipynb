```python
# stable algorithm gives nearly right answer to nearly question

# Example of Eigenvalues of a matrix

import scipy.linalg as la
import numpy as np

A = np.array([[1., 1000], [0, 1]])
B = np.array([[1, 1000], [0.001, 1]])

print(A)
print(B)
```

```python
wA, vrA = la.eig(A)
wB, vrB = la.eig(B)  # eigen PCA SVD  PCA background removal of a photo

wA, wB
```
